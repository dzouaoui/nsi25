2. Compléter la fonction  en python ci-dessous pour traduire l'algorithme proposé.

```python
def est_present(c,s):
    """
    renvoie True si le caractère c est présent dans s.
    renvoie False sinon.
    
    :param c: (str) le caractère à chercher
    :param s: (str) la chaine à tester
    :return: (bool)
```    

Dans l'exécution, on doit obtenir :

```python
    >>> est_present('t','python')
    True
    >>> est_present('z',"je ne trouve pas")
    False
    >>> est_present('a','')
    False
```


# Exercice 9 : niveau intermédiaire

Prendre un nombre entier:
- Si ce nombre est pair, le diviser par 2.
- Si ce nombre est impair, prendre le triple et ajouter 1.
On obtient un nouveau nombre entier et on recommence.

La conjecture de Syracuse dit qu’à la fin, on obtient toujours 1.
Vous pouvez consulter 
[wikipedia](https://fr.wikipedia.org/wiki/Conjecture_de_Syracuse)  ou [la vidéo](https://www.youtube.com/watch?v=aRe4ARtQiJY)

On peut définir aussi :

a. Le **temps de vol** de la suite est le plus petit nombre d'éléments de la suite pour arriver à 1.

b. L'altitude maximale de la suite est le plus grand élément de la suite.

**Exemple :** Pour n = 3, la suite de syracuse est:

3 - 10 - 5 - 16 - 8 - 4 - 2 - 1

Le temps du vol pour 3 est de 7 et son altitude maximale est de 16.

1. Ecrire une fonction _syracuse_ qui prend comme paramètre un entier naturel non nul et qui affiche la suite de syracuse.
2. Compléter la fonction _syracuse_ pour afficher la durée de vol et l'altitude de la suite.
